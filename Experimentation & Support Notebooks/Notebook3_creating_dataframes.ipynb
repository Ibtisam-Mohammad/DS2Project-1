```python
# renaming the columns

# concatenating all the headers
df.columns = df.columns.get_level_values(0) + '*' + df.columns.get_level_values(1)+ '*' + df.columns.get_level_values(2)
# Removing unnecessary parts of the headers
cols = []
for i in df.columns:
  k = i.split('*')
  k_imp = [s for s in k if not s.startswith('Unnamed:')]
  k_imp = [x.replace('\n','') for x in k_imp]
  k_fin = ': '.join(k_imp)
  cols.append(k_fin)
df.columns=cols
```
